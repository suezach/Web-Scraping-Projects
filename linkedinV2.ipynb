1. Use linkedin website, to get all the job details for all jobs listed.
2. The linkedin website new jobs load up as we keep scrolling down until all the 1000+ products are loaded up.
    This is called infinite scrolling.
3. Our project, we will get past infinite scrolling to get the job details of all the 1000+ jobs.
4. For this, we will use the combination of Selenium, and BeautifulSoup.
    Selenium: for infinite scrolling
    BeautifulSoup: to get the product details.


```python
# Import Libraries
#BeautifulSoup
from bs4 import BeautifulSoup
import pandas as pd
import requests

#Selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains # import library to scroll to specific element 
import time
```

```python
# Get the chrome driver and put it in the variable driver

driver = webdriver.Chrome('/Users/zachariah/Downloads/chromedriver')
```

```python
# get the url of job search page

driver.get('https://www.linkedin.com/jobs/jobs-in-kennett-square-pa?trk=homepage-basic_guest_nav_menu_jobs&position=1&pageNum=0')
```

```python
# INFINITE SCROLLING ON THIS PAGE

# 1. GET HEIGHT OF THE CURRENT PAGE

last_height = driver.execute_script('return document.body.scrollHeight')
last_height
```

```python
# universal Code to self scroll to the very last page of the items.

while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    #break
    time.sleep(2)
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == last_height:
        break
    last_height = new_height
```

```python
last_height
```

```python
# to get the html corresponding to each product
# frist we need to get the html of the page using BeautifulSoup 
# since we are using selenium we don't have to use the request library, instead we use the following code

soup = BeautifulSoup(driver.page_source, 'lxml')
```

```python
# create a dataframe

df2 = pd.DataFrame({'Job Title':[''], 'company' : [''],'Location':[''], 'Status':[''], 'Posted':[''],'Link':['']})
```

```python
# finding all li tags in ul and printing the text within it
data1 = soup.find_all('ul', class_ ='jobs-search__results-list')[0]

for job in data1.find_all('li'):
    try:
        link = job.find('a',class_='base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]').get('href')
        title = job.find('h3',class_='base-search-card__title').text.strip()
        company = job.find('a',class_='hidden-nested-link').text.strip()
        location = job.find('span',class_='job-search-card__location').text.strip()
        status = job.find('div', class_='result-benefits').text.strip()
        
        posted = job.find('time',class_='job-search-card__listdate').text.strip()
        df2 = df2.append({'Job Title':title, 'company' : company,'Location':location,'Status':status, 'Posted':posted,'Link':link}, ignore_index = True)
        #pass
    except:
        pass
        
```
