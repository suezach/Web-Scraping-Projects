CODING EXERCISE TO SCRAP THE NFL STANDINGS (REGULAR SEASON) 2021

GOAL : TO REPLICATE THE NFL STANDING TABLE

WEBSITE: https://www.nfl.com/standings/league/2021/REG

```python
# import libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Get the url of the Worldometer webpage

url = 'https://www.nfl.com/standings/league/2021/REG'
```

```python
# Check to see if we can access the HTML and put the output into variable page

page = requests.get(url)
page
```

```python
#Grab the HTML as text in page.text, amd lxml parser puts the text back into html format for python.
# HTML of webpage is now in Python and in the 'soup' variable.

soup = BeautifulSoup(page.text, 'lxml')
#soup
```

```python
# Extracting the HTML for the table that we are scrapping

table = soup.find('table', {'summary':'Standings - Detailed View'})
```

```python
table
```

```python
# First thing to do is to get the column headers

table.find_all('th')
```

```python
# we want only strings of the column headers 

headers = []
for i in table.find_all('th'):
    col_header = i.text
    headers.append(col_header)
```

```python
headers
```

```python
# Create a dataframe with headers as column names

df = pd.DataFrame(columns = headers)
```

```python
df
```

```python
# First get all tags that correspond to the rows in table on webpage
# Each row corresponds to 'tr' tag and data within it corresponds to 'td' tag

# Use find_all to get 'tr' tags

table.find_all('tr')
```

```python
#Since the first tr tags contains the headers, we don't want that, so we index it out

table.find_all('tr')[1:]
```

```python
# use for loop to loop through it
# To get the data from the web page into the python


for j in table.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [tr.text for tr in row_data]
    length = len(df)
    df.loc[length] = row
```
