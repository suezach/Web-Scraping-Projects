```python
df = pd.DataFrame({'Title':[''], 'Company':[''], 'Apply':[''], 'City':[''],'Posted':['']})
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
#postings = soup.find_all('ul', class_ ='reusable-search__entity-result-list list-style-none')
data = soup.find_all('ul', class_ ='reusable-search__entity-result-list list-style-none')[0]

# finding all li tags in ul and printing the text within it

        
for job in data.find_all('li'):
    title = job.find('span', class_ ='entity-result__title-text t-16').text.strip()
    company = job.find('div', class_ = 'entity-result__primary-subtitle t-14 t-black t-normal').text.strip()
    link = job.find('a', class_ ='app-aware-link').get('href')
    full_link =  'https://linkedin.com' + link
    city = job.find('div',class_ = 'entity-result__secondary-subtitle t-14 t-normal').text.strip()
    posted = job.find('div', 'display-flex align-items-center mt1').text.strip()
    df = df.append({'Title':title, 'Company':company, 'Apply':full_link, 'City':city,'Posted':posted}, ignore_index = True)
   
    
```
