SCRAPING CARPAGES ON www.carpages.com

Scrapping through each posting

1. Get the Link of each posting
2. Get the name of each car
3. Get the price of each car
4. Get the color of each car

Do this for 10-15 pages 


```python
# import libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Get the url of the website that needs to be scrapped
url ='https://www.carpages.ca/used-cars/search/?num_results=50&fueltype_id%5B0%5D=3&fueltype_id%5B1%5D=7&p=1'
```

```python
# Find if we can scrap this website
page = requests.get(url)
page
```

```python
# crerate variable soup to store the extracted html in python
soup = BeautifulSoup(page.text, 'lxml')
#soup
```

```python
df = pd.DataFrame({'Link':[''], 'Name':[''], 'Price':[''], 'Color':['']})
```

```python
counter = 0
while counter < 10:
    postings = soup.find_all('div', class_ = 'media soft push-none rule')
    
    for post in postings:
        link = post.find('a', class_ = 'media__img media__img--thumb')
        link_full = 'http://www.carpages.ca' + link.get('href')
        name = post.find('h4', class_='hN').text.strip()
        price = post.find('strong', class_ = 'delta').text.strip()
        color = post.find_all('div', class_ = 'grey l-column l-column--small-6 l-column--medium-4')[1].text.strip()
        df = df.append({'Link':link_full, 'Name':name, 'Price':price, 'Color':color}, ignore_index = True)
    
    next_p = soup.find('a', class_ = 'nextprev').get('href')
    next_page = 'http://www.carpages.ca' +next_p
    page = requests.get(next_page)
    soup = BeautifulSoup(page.text,'lxml')
    
    counter += 1
    
```
