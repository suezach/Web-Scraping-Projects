CHALLENGE: TO SCRAP MULTIPLE PAGES AT ONCE AND GET DATA FROM EACH PAGE

WEBSITE : www.airbnb.com

Scrap through every posting and get the following for each postings on all the pages.

1. name 
2. description 
3. price 
4. ratings


```python
# import libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Getting the HTML from a website

url= 'https://www.airbnb.com/s/Honolulu--Oahu--Hawaii--United-States/homes?tab_id=home_tab&flexible_trip_lengths%5B%5D=one_week&refinement_paths%5B%5D=%2Fhomes&place_id=ChIJTUbDjDsYAHwRbJen81_1KEs&query=Honolulu%2C%20Oahu%2C%20HI&date_picker_type=calendar&search_type=unknown'
```

```python
# Check to see if we can access the HTML and put the output into variable page

page = requests.get(url)
page
```

```python
#Grab the HTML as text in page.text, amd lxml parser puts the text back into html format for python.
# HTML of webpage is now in Python and in the 'soup' variable.

soup = BeautifulSoup(page.text, 'lxml')
#soup
```

```python
df = pd.DataFrame({'Link':[''],'Name':[''], 'Desc':[''], 'Price':[''], 'Ratings':['']})
```

```python
# we will use a loop to loop through all the pages

# on last page there is no arrow, so the loop will stop
counter = 0
while counter < 3:
    
    postings = soup.find_all('div', class_ = 'c4mnd7m dir dir-ltr') # get the html for each posting
    for post in postings:   # for loop to go through each posting and grab data from each one
        try:
            link = post.find('a', class_ ='rfexzly dir dir-ltr').get('href')
            link_full = 'https://www.airbnb.com' + link
            name= post.find('div', class_ ='t1jojoys dir dir-ltr').text
            desc = post.find('span', class_ = 'tjbvqj3 dir dir-ltr').text
            price = post.find('span', class_ = 'a8jt5op dir dir-ltr').text
            rating = post.find('span', class_ = 'ru0q88m dir dir-ltr').text
        except:
            pass
        df = df.append({'Link':link_full,'Name':name, 'Desc':desc, 'Price':price, 'Ratings':rating}, ignore_index = True)
    
    next_pages = soup.find('a', {'aria-label':'Next'})
    next_pages_full = 'https://www.airbnb.com' + next_pages.get('href')
    pages = requests.get(next_pages_full)
    soups = BeautifulSoup(pages.text,'lxml')
    
counter += 1 
```
